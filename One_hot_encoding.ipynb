```python
import pandas as pd
import numpy as np
```

```python
raw_data = {'food_name': ['Apple', 'Chicken', 'Broccoli', 'Chocolat'],'Calories': [95, 231, 50, 549]}
dataset = pd.DataFrame(raw_data)
dataset.head()
```

Import Libraries: Importeert de benodigde libraries: pandas voor data-analyse en numpy voor numerieke operaties.

Gegevensinvoer:

Maakt een dictionary raw_data met voedselnamen en calorieën.

Zet deze om in een pandas DataFrame.

dataset.head() toont de eerste paar rijen van de DataFrame om te controleren of de gegevens correct zijn ingelezen.

Resultaat:

```python
# One-hot encoding


dataset = pd.concat([dataset,pd.get_dummies(dataset['food_name'], prefix='food')],axis=1)
dataset.drop(['food_name'],axis=1, inplace=True)
dataset.head()
```

Uitleg:
One-Hot Encoding:

Gebruikt pd.get_dummies() om kolommen te maken voor elke unieke waarde in food_name.

prefix='food' voegt een voorvoegsel toe aan de gegenereerde kolommen.

Samenvoegen van DataFrames:

De gegenereerde dummy-kolommen worden samengevoegd met de originele DataFrame.

Kolom Verwijderen:

Verwijdert de oorspronkelijke food_name kolom omdat deze nu overbodig is.

Resultaat

```python
# alternative way

from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer

raw_data = {'food_name': ['Apple', 'Chicken', 'Broccoli', 'Chocolat'],'Calories': [95, 231, 50, 549]}
dataset = pd.DataFrame(raw_data)


ohe = OneHotEncoder()
imp = SimpleImputer()

ct = make_column_transformer((ohe,['food_name']),(imp,['Calories']),remainder='passthrough',verbose_feature_names_out=False)
encoded = ct.fit_transform(dataset)
enc_features = ct.get_feature_names_out()
dataset_enc = pd.DataFrame(data= encoded, columns=enc_features)

dataset_enc.head()
```

Uitleg:
OneHotEncoder: Om categorische gegevens (voedselnaam) te coderen.

SimpleImputer: Voor het eventueel opvullen van ontbrekende waarden (hier eigenlijk niet nodig, maar toegevoegd voor algemene bruikbaarheid).

make_column_transformer:

Combineert de encodering en imputatie in één transformer.

remainder='passthrough': Laat de niet-getransformeerde kolommen zoals ze zijn.

verbose_feature_names_out=False: Houdt de kolomnamen compact (zonder prefix).

Transformatie:
fit_transform: Past de transformator toe op de originele DataFrame.

get_feature_names_out: Verkrijgt de namen van de gegenereerde kolomme
